```python
import numpy as np
import math
from matplotlib import pyplot
from random import randrange
```

## Definicion de funciones

```python
def logSigmoid(a):
    b = 1/(1+math.exp(-a))
    return b
```

```python
def purelin(a):
    if a < 0:
        b = 0
    else:
        b = a
    return b
```

```python
def g(x):
    return 1+math.sin((math.pi/4)*x)
```

```python
#Primera derivada
def f1(a):
    return (1-a)*(a)
```

```python
#Segunda derivada
def f2(x):
    return 1
```

## Forward Propagation

```python
#Forward Propagation
def forwardPropagation():
    u = w1_0*p + b1_0
    a1 = np.matrix([[logSigmoid(u[0])],[logSigmoid(u[1])]])
    u = w2_0 * a1 + b2_0
    a2 = purelin(u)
```

```python
#Calculando error
def error():
    e = g(p) - a2
```
